```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("multiinput_linear_regression.csv")
```

```python
df.head()
```

```python
# remaning the column for better understanding
df.columns = ["A" , "B" , "C"]
```

```python
# now we plot the scatter plot between the data
sns.scatterplot(x = "A" , y = "B" , hue='C' , data = df)
plt.show()
```

```python
# draw the pair plot for better understaning of the data
sns.pairplot(df)
plt.show()
```

from the above graph we have seen that this is the linear data with multiple input

```python
# sepaarting the data for X and Y
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3,random_state=41)
```

```python
# prompt: generate the linear regression model , train and test the data and return the accuracy report

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Create and train the model
model = LinearRegression()
model.fit(x_train, y_train)

# Make predictions
y_pred = model.predict(x_test)

# Evaluate the model
accuracy = r2_score(y_test, y_pred)
print("Accuracy:", accuracy*100)
```

```python
print("testing accuracy" , model.score(x_test,y_test)*100)
print("training accuracy" , model.score(x_train,y_train)*100)
```

```python
# prompt: return the cost function values

from sklearn.metrics import mean_squared_error
# Calculate the cost function (mean squared error)
cost = mean_squared_error(y_test, y_pred)
print("Cost Function Value:", cost)
```

```python
# prompt: draw the graph after the model training that show the best fit line

import matplotlib.pyplot as plt
# Plot the best fit line
plt.figure(figsize=(5, 4))
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values")

# Add the best fit line
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')

plt.show()
```

```python
print("this is the graph after the testing")
generate_post_testing_plot = 'yes'
if generate_post_testing_plot == 'yes':
    plt.figure(figsize=(5, 4))
    plt.scatter(x_test['A'], y_test, label='Actual Salary')
    plt.plot(x_test['A'], y_pred, color='red', label='Predicted Salary')
    plt.xlabel('Years of Experience')
    plt.ylabel('Salary')
    plt.title('Linear Regression Model - Actual vs Predicted')
    plt.legend()
    plt.show()
```

```python
print("this is the graph before the training : ")
generate_pre_training_plot = 'yes'
if generate_pre_training_plot == 'yes':
    plt.figure(figsize=(8, 6))
    plt.scatter(x_train['A'], y_train, label='Training Data')
    plt.scatter(x_test['A'], y_test, label='Testing Data')
    plt.xlabel('Years of Experience')
    plt.ylabel('Salary')
    plt.title('Data Distribution Before Model Training')
    plt.legend()
    plt.show()
```
