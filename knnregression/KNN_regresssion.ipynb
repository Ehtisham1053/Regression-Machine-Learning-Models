```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.drop("date" , inplace=True , axis=1)
```

```python
df['country'].replace({'USA' : 1} , inplace=True)
```

```python
df['country'].unique()
```

```python
# prompt: prompt for ordinal encoding for the statezip column

from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()
df["statezip_encoded"] = ordinal_encoder.fit_transform(df[["statezip"]])
```

```python
df.drop(columns=["statezip_encoded"] , inplace=True)
```

```python
df.head()
```

```python
x = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
# prompt: scale the x data using standardscaler

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state=42)
```

```python
# prompt: generate the KNN regression model ,train and test the data return the accuracy report

from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Create KNN regressor
knn = KNeighborsRegressor(n_neighbors=5)  # You can adjust n_neighbors

# Train the model
knn.fit(x_train, y_train)

# Make predictions
y_pred = knn.predict(x_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```
